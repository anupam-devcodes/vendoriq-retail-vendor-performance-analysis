```python
import pandas as pd
from pathlib import Path
from sqlalchemy import create_engine

current_path = Path.cwd()

if current_path.name == "notebooks":
    BASE_DIR = current_path.parent
else:
    BASE_DIR = current_path

DATABASE_PATH = BASE_DIR / "database" / "vendor_iq.db"

engine = create_engine(f"sqlite:///{DATABASE_PATH.as_posix()}", echo=False)

print("Connected to:", DATABASE_PATH)
```

```python
vendor_match_query = """
WITH sales_vendors AS (
    SELECT DISTINCT VendorNo AS VendorNumber
    FROM sales
),
purchase_vendors AS (
    SELECT DISTINCT VendorNumber
    FROM purchases
)

SELECT 
    'Vendors in sales' AS category,
    COUNT(*) AS vendor_count
FROM sales_vendors

UNION ALL

SELECT 
    'Vendors in purchases' AS category,
    COUNT(*) AS vendor_count
FROM purchase_vendors

UNION ALL

SELECT 
    'Vendors in both sales and purchases' AS category,
    COUNT(*) AS vendor_count
FROM sales_vendors
INNER JOIN purchase_vendors
    ON sales_vendors.VendorNumber = purchase_vendors.VendorNumber;
"""

vendor_match = pd.read_sql_query(vendor_match_query, engine)

vendor_match
```

```python
unmatched_vendors_query = """
WITH sales_vendors AS (
    SELECT DISTINCT 
        VendorNo AS VendorNumber,
        VendorName
    FROM sales
),
purchase_vendors AS (
    SELECT DISTINCT 
        VendorNumber,
        VendorName
    FROM purchases
)

SELECT 
    'Only in sales' AS match_status,
    sales_vendors.VendorNumber,
    sales_vendors.VendorName
FROM sales_vendors
LEFT JOIN purchase_vendors
    ON sales_vendors.VendorNumber = purchase_vendors.VendorNumber
WHERE purchase_vendors.VendorNumber IS NULL

UNION ALL

SELECT 
    'Only in purchases' AS match_status,
    purchase_vendors.VendorNumber,
    purchase_vendors.VendorName
FROM purchase_vendors
LEFT JOIN sales_vendors
    ON purchase_vendors.VendorNumber = sales_vendors.VendorNumber
WHERE sales_vendors.VendorNumber IS NULL;
"""

unmatched_vendors = pd.read_sql_query(unmatched_vendors_query, engine)

unmatched_vendors
```

```python
brand_match_query = """
WITH sales_brands AS (
    SELECT DISTINCT Brand
    FROM sales
),
purchase_brands AS (
    SELECT DISTINCT Brand
    FROM purchases
),
price_brands AS (
    SELECT DISTINCT Brand
    FROM purchase_prices
)

SELECT 
    'Brands in sales' AS category,
    COUNT(*) AS brand_count
FROM sales_brands

UNION ALL

SELECT 
    'Brands in purchases' AS category,
    COUNT(*) AS brand_count
FROM purchase_brands

UNION ALL

SELECT 
    'Brands in purchase_prices' AS category,
    COUNT(*) AS brand_count
FROM price_brands

UNION ALL

SELECT 
    'Brands in both sales and purchases' AS category,
    COUNT(*) AS brand_count
FROM sales_brands
INNER JOIN purchase_brands
    ON sales_brands.Brand = purchase_brands.Brand

UNION ALL

SELECT 
    'Brands in sales, purchases, and purchase_prices' AS category,
    COUNT(*) AS brand_count
FROM sales_brands
INNER JOIN purchase_brands
    ON sales_brands.Brand = purchase_brands.Brand
INNER JOIN price_brands
    ON sales_brands.Brand = price_brands.Brand;
"""

brand_match = pd.read_sql_query(brand_match_query, engine)

brand_match
```

```python
sales_summary_query = """
SELECT
    VendorNo AS VendorNumber,
    VendorName,
    Brand,
    Description,
    SUM(SalesQuantity) AS total_sales_quantity,
    SUM(SalesDollars) AS total_sales_dollars,
    AVG(SalesPrice) AS avg_sales_price
FROM sales
GROUP BY 
    VendorNo,
    VendorName,
    Brand,
    Description
ORDER BY total_sales_dollars DESC;
"""

sales_summary = pd.read_sql_query(sales_summary_query, engine)

sales_summary.head()
```

```python
purchase_summary_query = """
SELECT
    VendorNumber,
    VendorName,
    Brand,
    Description,
    SUM(Quantity) AS total_purchase_quantity,
    SUM(Dollars) AS total_purchase_dollars,
    AVG(PurchasePrice) AS avg_purchase_price
FROM purchases
GROUP BY
    VendorNumber,
    VendorName,
    Brand,
    Description
ORDER BY total_purchase_dollars DESC;
"""

purchase_summary = pd.read_sql_query(purchase_summary_query, engine)

purchase_summary.head()
```

```python
vendor_brand_summary_query = """
WITH sales_summary AS (
    SELECT
        VendorNo AS VendorNumber,
        VendorName,
        Brand,
        Description,
        SUM(SalesQuantity) AS total_sales_quantity,
        SUM(SalesDollars) AS total_sales_dollars,
        AVG(SalesPrice) AS avg_sales_price
    FROM sales
    GROUP BY 
        VendorNo,
        VendorName,
        Brand,
        Description
),

purchase_summary AS (
    SELECT
        VendorNumber,
        VendorName,
        Brand,
        Description,
        SUM(Quantity) AS total_purchase_quantity,
        SUM(Dollars) AS total_purchase_dollars,
        AVG(PurchasePrice) AS avg_purchase_price
    FROM purchases
    GROUP BY
        VendorNumber,
        VendorName,
        Brand,
        Description
)

SELECT
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    p.Description,
    p.total_purchase_quantity,
    p.total_purchase_dollars,
    p.avg_purchase_price,
    s.total_sales_quantity,
    s.total_sales_dollars,
    s.avg_sales_price,
    s.total_sales_dollars - p.total_purchase_dollars AS gross_profit,
    ROUND(
        (s.total_sales_dollars - p.total_purchase_dollars) * 100.0 
        / s.total_sales_dollars, 
        2
    ) AS profit_margin_percent,
    p.total_purchase_quantity - s.total_sales_quantity AS unsold_quantity
FROM purchase_summary p
INNER JOIN sales_summary s
    ON p.VendorNumber = s.VendorNumber
    AND p.Brand = s.Brand
ORDER BY gross_profit DESC;
"""

vendor_brand_summary = pd.read_sql_query(vendor_brand_summary_query, engine)

vendor_brand_summary.head()
```

```python
create_vendor_performance_summary_query = """
DROP TABLE IF EXISTS vendor_performance_summary;

CREATE TABLE vendor_performance_summary AS
WITH sales_summary AS (
    SELECT
        VendorNo AS VendorNumber,
        VendorName,
        Brand,
        Description,
        SUM(SalesQuantity) AS total_sales_quantity,
        SUM(SalesDollars) AS total_sales_dollars,
        AVG(SalesPrice) AS avg_sales_price
    FROM sales
    GROUP BY 
        VendorNo,
        VendorName,
        Brand,
        Description
),

purchase_summary AS (
    SELECT
        VendorNumber,
        VendorName,
        Brand,
        Description,
        SUM(Quantity) AS total_purchase_quantity,
        SUM(Dollars) AS total_purchase_dollars,
        AVG(PurchasePrice) AS avg_purchase_price
    FROM purchases
    GROUP BY
        VendorNumber,
        VendorName,
        Brand,
        Description
)

SELECT
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    p.Description,

    p.total_purchase_quantity,
    p.total_purchase_dollars,
    ROUND(p.avg_purchase_price, 2) AS avg_purchase_price,

    s.total_sales_quantity,
    s.total_sales_dollars,
    ROUND(s.avg_sales_price, 2) AS avg_sales_price,

    ROUND(s.total_sales_dollars - p.total_purchase_dollars, 2) AS gross_profit,

    ROUND(
        (s.total_sales_dollars - p.total_purchase_dollars) * 100.0 
        / NULLIF(s.total_sales_dollars, 0), 
        2
    ) AS profit_margin_percent,

    p.total_purchase_quantity - s.total_sales_quantity AS unsold_quantity,

    ROUND(
        s.total_sales_quantity * 1.0 / NULLIF(p.total_purchase_quantity, 0),
        2
    ) AS sales_to_purchase_qty_ratio,

    ROUND(
        s.total_sales_dollars * 1.0 / NULLIF(p.total_purchase_dollars, 0),
        2
    ) AS sales_to_purchase_value_ratio

FROM purchase_summary p
INNER JOIN sales_summary s
    ON p.VendorNumber = s.VendorNumber
    AND p.Brand = s.Brand;
"""
```

```python
with engine.begin() as conn:
    for statement in create_vendor_performance_summary_query.strip().split(";"):
        if statement.strip():
            conn.execute(text(statement))
```
